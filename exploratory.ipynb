```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('yelp_boston.csv')
df.head()
```

```python
categories = ['breakfast_brunch', 'burgers', 'sandwiches', 'mexican', 'japanese', 'italian', 'pizza', 'seafood', 'raw_food', 'newamerican', 
              'mediterranean', 'vegan', 'gluten_free', 'spanish', 'wine_bars', 'tapasmallplates', 'salad', 'juicebars', 'vegetarian', 'indpak', 
              'belgian', 'tradamerican', 'pubs', 'korean', 'asianfusion', 'lounges', 'coffee', 'delis', 'grocery', 'bakeries', 'gastropubs', 
              'bars', 'cocktailbars', 'desserts', 'french', 'steak', 'cheese', 'bubbletea', 'vietnamese', 'catering', 
              'cheesesteaks', 'turkish', 'falafel', 'pakistani', 'bangladeshi', 'himalayan', 'halal', 'hotdogs', 'sushi', 'hotpot', 'mongolian', 
              'dimsum', 'cantonese', 'foodtrucks', 'latin', 'thai', 'icecream', 'donuts', 'cupcakes', 'greek', 'ethnicmarkets', 
              'armenian', 'mideastern']

for c in categories:
    df[c]= 0    
    df.loc[df['categories_json'].str.contains(c, regex=False), c] = 1

df.head()
```

```python
df2 = df[['name', 'breakfast_brunch', 'burgers', 'sandwiches', 'mexican', 'japanese', 'italian', 'pizza', 'seafood', 'raw_food', 'newamerican', 
              'mediterranean', 'vegan', 'gluten_free', 'spanish', 'wine_bars', 'tapasmallplates', 'salad', 'juicebars', 'vegetarian', 'indpak', 
              'belgian', 'tradamerican', 'pubs', 'korean', 'asianfusion', 'lounges', 'coffee', 'delis', 'grocery', 'bakeries', 'gastropubs', 
              'bars', 'cocktailbars', 'desserts', 'french', 'steak', 'cheese', 'bubbletea', 'vietnamese', 'catering', 
              'cheesesteaks', 'turkish', 'falafel', 'pakistani', 'bangladeshi', 'himalayan', 'halal', 'hotdogs', 'sushi', 'hotpot', 'mongolian', 
              'dimsum', 'cantonese', 'foodtrucks', 'latin', 'thai', 'icecream', 'donuts', 'cupcakes', 'greek', 'ethnicmarkets', 
              'armenian', 'mideastern']]
df3 = pd.DataFrame()
for column in categories:
    df3[column] = df2[[column, 'name']].groupby(column).agg({'name': 'sum', 'name': lambda x: '; '.join(x)}).loc[1]

df3 # a dataframe with a string list (separated by semicolons) of all of the restaurants within that category
```
